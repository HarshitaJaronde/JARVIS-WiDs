```python
import pandas as pd
```

```python
df = pd.read_csv("C:\\Users\\HARSHITA\\Downloads\\Dataset.csv")
df
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train,y_test = train_test_split(df.Speech, df.Intent, test_size = 0.1, random_state=21)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
clf = Pipeline([('vectorizer', CountVectorizer()),('nb',LogisticRegression())])
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
clf.score(X_test,y_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
import pickle 
cr = classification_report(y_test, y_pred)
with open('cr_pickle','wb') as f:
    pickle.dump(cr,f)
```
